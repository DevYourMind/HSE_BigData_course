```python
title_basics_csv = spark.read.csv("title.basics.tsv.gz", sep='\\t', nullValue='\\N', header=True)
```

```python
title_basics_csv.createOrReplaceTempView("title_basics_csv")
```

```python
sql("SELECT * FROM title_basics_csv LIMIT 10").show()
```

```python
title_principals_csv = spark.read.csv("title.principals.tsv.gz", sep='\\t', nullValue='\\N', header=True)
title_principals_csv.createOrReplaceTempView("title_principals_csv")

title_crew_csv = spark.read.csv("title.crew.tsv.gz", sep='\\t', nullValue='\\N', header=True)
title_crew_csv.createOrReplaceTempView("title_crew_csv")

title_episode_csv = spark.read.csv("title.episode.tsv.gz", sep='\\t', nullValue='\\N', header=True)
title_episode_csv.createOrReplaceTempView("title_episode_csv")

title_ratings_csv = spark.read.csv("title.ratings.tsv.gz", sep='\\t', nullValue='\\N', header=True)
title_ratings_csv.createOrReplaceTempView("title_ratings_csv")

title_akas_csv = spark.read.csv("title.akas.tsv.gz", sep='\\t', nullValue='\\N', header=True)
title_akas_csv.createOrReplaceTempView("title_akas_csv")

name_basics_csv = spark.read.csv("name.basics.tsv.gz", sep='\\t', nullValue='\\N', header=True)
name_basics_csv.createOrReplaceTempView("name_basics_csv")
```

```python
sql("SELECT DISTINCT explode(split(genres, '[,]', -1)) FROM title_basics_csv").show()
```

```python
sql("SELECT DISTINCT explode(split(genres, '[,]', -1))  FROM title_basics_csv ORDER BY 1").show()
```
