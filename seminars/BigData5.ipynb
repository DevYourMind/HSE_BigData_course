```python
import requests
header = requests.get('http://kdd.ics.uci.edu/databases/kddcup99/kddcup.names').text.split('\n')[1:-1]

types = [h.split(':')[1].strip(' .') for h in header]
header = [h.split(':')[0] for h in header]
```

```python
header
```

```python
data_with_header = df
for i, h in enumerate(header + ['tag']):
    data_with_header = data_with_header.withColumnRenamed('_c{}'.format(i), h)
data_with_header.limit(5).toPandas()
```

```python
data_with_header.show(5)
```

```python
data_with_header.printSchema()
```

```python
import pyspark.sql.functions as sf

def cast_if_continuous(col_name, t):
    if t == u'continuous':
        return sf.col(col_name).cast('float').alias(col_name)
    else:
        return sf.col(col_name)

data_with_types = data_with_header.select([cast_if_continuous(col_name, t) for col_name, t in zip (header, types)] + [sf.col("tag")])
```

```python
data_with_types.show()
```

```python
data_with_types.printSchema()
```

```python
data_with_types.select("tag", "protocol_type", 
                       ((sf.col("dst_host_serror_rate")+sf.col("dst_host_srv_serror_rate")+sf.col("dst_host_rerror_rate")+sf.col("dst_host_srv_rerror_rate"))/4)
                       .alias("mean_error_rate")).orderBy("mean_error_rate", ascending=False).show()
```

```python
data_with_types.groupby("tag", "protocol_type").agg(sf.count(sf.col("tag"))).show()
```

```python
'src_bytes'
```

```python
mean_src_bytes_by_tag_df = data_with_types.groupby("tag").agg(sf.mean("src_bytes").alias("mean"))

mean_src_bytes_by_tag = {r['tag']: r['mean'] for r in mean_src_bytes_by_tag_df.collect()}
assert mean_src_bytes_by_tag['teardrop.'] == 28
```

```python
data_with_types.createOrReplaceTempView("data")
```

```python
df_sql = ss.sql("select tag, protocol_type, count(tag) as count from data group by tag, protocol_type")
df_sql.show()
```

```python
df_sql.explain()
```

```python
@sf.udf
def reverse_string(val):
    return val[::-1]
```

```python
df_sql.select("tag", reverse_string("tag")).show()
```

```python
@sf.udf
def tag_protocol_type(tag, protocol_type):
  return f"{tag}_{protocol_type}"
```

```python
df_protocol_sql = data_with_types.groupby(tag_protocol_type("tag", "protocol_type").alias("key")).agg(sf.count("tag"))
df_protocol_sql.show()
```

```python
df_protocol_sql.explain()
```
